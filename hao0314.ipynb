```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib  inline
```

```python
iris=datasets.load_breast_cancer()
```

```python
iris
```

```python
x= pd.DataFrame(iris['data'], columns=iris['feature_names'])
print("target_names: "+str(iris['target_names']))
y= pd.DataFrame(iris['target'], columns=['target'])
```

```python
iris_data = pd.concat([x,y], axis=1)
iris_data.head(3)
```

```python
iris_2class = iris_data.loc[0:99,:]
iris_2class
```

```python
import seaborn as sns
plt.figure(figsize=(30,8))
sns.heatmap(round(iris_2class.corr(),4),annot=True,vmin=-1,vmax=1,cmap='coolwarm')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test,=train_test_split(
    iris_2class[['sepal length (cm)','petal length (cm)']],
    iris_2class[['target']],test_size=0.2,random_state=0)

```

```python
X_train,X_valid, y_train, y_valid = train_test_split( X_train, y_train , test_size=0.25, random_state=0)
```

```python
len(X_train),len(X_valid),len(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
x_train_std = sc.transform(X_train)
x_valid_std = sc.transform(X_valid)
x_test_std = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(x_train_std,y_train['target'].values)
```

```python
pred_knn = knn.predict(x_test_std)
pred_knn_prob = knn.predict_proba(x_test_std)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score, roc_curve

accuracy_knn = accuracy_score(y_test, pred_knn)
return_knn = recall_score(y_test, pred_knn)
f1_score_knn = f1_score(y_test, pred_knn)
print('accuracy: ', accuracy_knn)
print("recall score: ", return_knn)
print("f1 score: ", f1_score_knn)
```

```python
cm = confusion_matrix(y_test, pred_knn)
sns.set(font_scale=1.3)
sns.heatmap(cm, annot=True, cmap='Blues')
```

```python
fpr,tpr,threshold=roc_curve(y_test,pred_knn_prob[:,1])
auc=roc_auc_score(y_test,pred_knn_prob[:,1])
print("AUC",auc)
```

```python
plt.plot(figsize=(5,5))
plt.plot(fpr,tpr,label="auc={0}".format(auc))
plt.axis("square")
plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
plt.legend()
plt.xlim(-0.1,1.1)
plt.ylim(-0.1,1.1)
plt.title("ROC curve")
```
